```python
import numpy as np
import plotly.graph_objects as go

# Generación de datos
x1 = np.random.randn(1, 100)
x1 = x1 - np.min(x1)
y1 = np.random.randn(1, 100)
y1 = y1 - np.min(y1)

x2 = np.random.randn(1, 100)
x2 = x2 - np.min(x2) + 4
y2 = np.random.randn(1, 100)
y2 = y2 - np.min(y2) + 2

# Definición de trazas
trace1 = go.Scatter(
    x=x1.flatten(),
    y=y1.flatten(),
    text='Cluster 1',
    mode='markers',
    visible=True
)

trace2 = go.Scatter(
    x=x2.flatten(),
    y=y2.flatten(),
    text='Cluster 2',
    mode='markers',
    visible=True
)

# Definición de diseño
layout = go.Layout(
    title='Clusters',
    updatemenus=[{
        'showactive': True,
        'direction': 'down',
        'buttons': [
            {
                'args': [{'visible': [True, True]}],
                'label': "Clusters",
                'method': "update"
            },
            {
                'args': [{'visible': [True, False]}],
                'label': "Cluster 1",
                'method': "update"
            }
        ]
    }]
)

# Creación de la figura
fig_btns = go.Figure(data=[trace1, trace2], layout=layout)
fig_btns.show()
```
